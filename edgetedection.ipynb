```python
#import image from the computer
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
import imageio
from findpeaks import findpeaks
import scipy as sp

volume_properties= []
```

```python

# Read the original image
img = imageio.imread("photos/lego_sample.png")

# Display original image
plt.imshow( img)
plt.title('Original Image')

 
# Convert to graycsale
img_gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
# Blur the image for better edge detection
img_blur = cv.GaussianBlur(img_gray, (3,3), 0) 
plt.figure()
i=0
for tr1 in range(0,250,50):
    for tr2 in range(0,250,50):
        # Canny Edge Detection
        edges = cv.Canny(image=img_blur, threshold1=tr1, threshold2=tr2) 
        # Display Canny Edge Detection Image
        plt.subplot(5, 5, i+1)
        i=i+1
        plt.imshow(edges)
        plt.title(str(tr1) +' '+ str( tr2))
        #dont show axis
        plt.xticks([])
        plt.yticks([])
plt.show()

    


```
